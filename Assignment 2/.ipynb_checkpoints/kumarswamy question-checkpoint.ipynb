```python
import numpy as np
```

```python
import seaborn as sns
```

```python
x=np.random.uniform(0,1,100)
```

```python
inverseCDF= [0] * 100
```

```python
type(inverseCDF)
```

```python
alpha=2
```

```python
beta=5
```

```python
inverseCDF=(1-((1-x[1])**(1/beta)))**(1/alpha)
```

```python
inverseCDF
```

```python
for i in range(0,100):
    inverseCDF[i]=(1-((1-x[i])**(1/beta)))**(1/alpha)
```

```python
np.mean(inverseCDF)
```

```python
np.var(inverseCDF)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(inverseCDF,normed=True,bins=10)
```
