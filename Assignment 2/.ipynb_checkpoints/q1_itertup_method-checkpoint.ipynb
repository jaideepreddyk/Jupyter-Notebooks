```python
class Primarycard:
    primaryCardLimitDict = {}
    def __init__(self,cust_id,card_no,acc_no,card_limit,current_bal):
        self.__cust_id=cust_id
        self.__card_no=card_no
        self.__acc_no=acc_no
        self.__card_limit=card_limit
        self.current_bal=current_bal
        #self.total_bal=total_bal
        
        
    def cust_id(self):
        return self.__cust_id
    
    def card_no(self):
        return self.__card_no
    
    def acc_no(self):
        return self.__acc_no
    
    def card_limit(self):
        return self.__card_limit
    
    def current_bal(self):
        return self.current_bal
    
   # def total_bal(self):
    #    return __self.total_bal
        

    def current_balance(self,trans_amount):
                self.current_bal=self.current_bal+trans_amount
                if not self.card_no in Primarycard.primaryCardLimitDict.keys(): 
                    Primarycard.primaryCardLimitDict[self.card_no] = self.current_bal
                else:
                    Primarycard.primaryCardLimitDict[self.card_no] = self.current_bal + Primarycard.primaryCardLimitDict[self.card_no]
                
                return self.current_bal
            
    def accountBalance(self,trans_amount,total_bal):
        total_bal[self.__acc_no] = total_bal[self.__acc_no]+trans_amount
        return total_bal
    
    def summingAllAccountno(self,total_bal):
        if not self.__acc_no in total_bal.keys():
            total_bal[self.__acc_no] = self.current_bal
        else:   
            total_bal[self.__acc_no] = total_bal[self.__acc_no]+self.current_bal
        return total_bal

    
class Secondarycard(Primarycard):
    secondaryCardLimitDict = {}
    def __init__(self,cust_id,card_no,acc_no,card_limit,current_bal):
        super().__init__(cust_id,card_no,acc_no,card_limit,current_bal)
    
    def current_balance(self,transaction_amt):
        self.current_bal=self.current_bal+transaction_amt
        if not self.card_no in Secondarycard.secondaryCardLimitDict.keys(): 
            Secondarycard.secondaryCardLimitDict[self.card_no] = self.current_bal
        else:
            Secondarycard.secondaryCardLimitDict[self.card_no] = self.current_bal+Secondarycard.secondaryCardLimitDict[self.card_no]
            return self.current_bal
        
    #def update_accountbalance(self,transaction_amt,total_bal):
        #total_bal[acc_no]=
    
    
```
