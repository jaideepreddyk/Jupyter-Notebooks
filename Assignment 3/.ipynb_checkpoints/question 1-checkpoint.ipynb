```python
import requests
import re
```

```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Presidents_of_the_United_States"
```

```python
def getPresidents(url):
    urlreq = requests.get(url)
    urltext = urlreq.text
    expression = re.compile(r'title="Presidency of+.+">\d\d\d\d+.')
    matches = expression.findall(urltext)
    nameonly = re.compile(r'Presidency of (.+?)"')
    List_of_Presidents=list()
    for i in matches:
        name=nameonly.findall(i)
        List_of_Presidents.append(name[0])
    return List_of_Presidents

```

```python
def getBirthdays():
    Birthday_List=list()
    for i in getPresidents(url):
        x=i.replace(" ", "_")     ## to replace whitespaces in the name to 'underscore' as it is the url format
        urlreq = requests.get("https://en.wikipedia.org/wiki/"+x)
        urltext = urlreq.text
        pattern = re.compile(r'\d\d\d\d+-+\d\d+-+\d\d')
        matches = pattern.findall(urltext)
        Birthday_List.append(matches[0])
    return pd.Series(getPresidents(url),Birthday_List)
    

```

```python
getPresidents(url)
```

```python
getBirthdays()
```

```python
name_birth_series=getBirthdays()
```

##### Plotting time series

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime
```

```python
dates=name_birth_series.keys()
```

```python
dates=np.array(dates)
```

```python
names=name_birth_series.get_values()
```

```python
dates = [datetime.strptime(ii, "%Y-%m-%d") for ii in dates]
```

```python
levels=list()
for i in range(len(dates)):
    x=dates[i].day
    levels.append(x)
    
    
```

```python
levels=np.array(levels)
```

```python
fig, ax = plt.subplots(figsize=(40, 15))

start = min(dates)
stop = max(dates)
ax.plot((start, stop), (0, 0), 'k', alpha=.5)

for ii, (iname, idate) in enumerate(zip(names, dates)):
    level = levels[ii % 6]
    vert = 'top' if level < 0 else 'bottom'

    ax.scatter(idate, 0, s=100, facecolor='w', edgecolor='k', zorder=9999)
    ax.plot((idate, idate), (0, level), c='r', alpha=.7)
    ax.text(idate, level, iname,
            horizontalalignment='right', verticalalignment=vert, fontsize=14,
            backgroundcolor=(1., 1., 1., .3))
ax.set(title="US Presidents Birth Timeline")

ax.get_xaxis().set_major_locator(mdates.MonthLocator(interval=36))  ## The interval on the baseline is take for 3 years
ax.get_xaxis().set_major_formatter(mdates.DateFormatter("%b %Y"))
fig.autofmt_xdate()

plt.setp((ax.get_yticklabels() + ax.get_yticklines() +
          list(ax.spines.values())), visible=False)
plt.show()
```
