```python
import pandas as pd
```

```python
df=pd.read_csv('chicago_crime_data_v3.csv')
```

```python
df_pid=pd.read_csv('primary_type_id.csv')
```

```python
final_df=pd.merge(df,df_pid,on=['Primary Type ID'])
```

```python
final_df.head()
```

## High Frequency crimes separation

```python
high_frequency_crimes=final_df.groupby('Primary Type').size().reset_index(name="Frequency")
```

```python
high_frequency_crimes=high_frequency_crimes[high_frequency_crimes['Frequency']>10000]
```

```python
high_frequency_crimes.index=range(1,12)
```

```python
high_frequency_crimes
```

```python
crime_date=final_df[['Primary Type','Date']]
```

```python
crime_date['Date']=pd.to_datetime(crime_date['Date'])
```

```python
crime_date['month_year'] = crime_date.Date.dt.to_period('M')
```

```python
crime_date=crime_date.drop('Date',axis=1)
```

```python
hf_crime_list=list(high_frequency_crimes['Primary Type'])
```

```python
high_crime_date=crime_date.loc[crime_date['Primary Type'].isin(hf_crime_list)]
```

```python
high_crime_date['Primary Type'].describe()
```

```python
high_crime_date.head()
```

```python
HF_crime_types=high_crime_date.groupby(['Primary Type','month_year']).size().reset_index(name="Frequency")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
hf_crime_list
```

```python
for i in hf_crime_list:
    x=high_crime_date[high_crime_date['Primary Type']==i]['month_year'].value_counts()
    x.plot.bar()
    plt.title(i)
    plt.show()
```

```python
import os,conda

conda_file_dir = conda.__file__
conda_dir = conda_file_dir.split('lib')[0]
proj_lib = os.path.join(os.path.join(conda_dir, 'share'), 'proj')
os.environ["PROJ_LIB"] = proj_lib

from mpl_toolkits.basemap import Basemap
```

```python
import numpy as np
```

```python
basemap_df = final_df[final_df['Primary Type'].isin(hf_crime_list)].iloc[:,np.r_[2,11,12]]
basemap_df = basemap_df.sort_values("Date").fillna(method='ffill').fillna(method='backfill')
```

```python
fig = plt.figure(figsize=[16,10])
m = Basemap(projection='merc',llcrnrlat=41.60,urcrnrlat=42.10,\
        llcrnrlon=-88.0,urcrnrlon=-87.50,lat_ts=20,resolution='c')
m.readshapefile('./geo_export_0e3da441-8fe8-4e11-9ca0-42ef75cab68e','chicago')

lons = list(basemap_df["Longitude"])
lats = list(basemap_df["Latitude"])
x,y = m(lons,lats)
plt.scatter(x,y,alpha=0.1,s=10)
plt.title("Chicago Cenus Tracts")
plt.show()
```
