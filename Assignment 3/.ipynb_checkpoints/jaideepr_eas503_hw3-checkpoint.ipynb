````python
import requests
import re
````

````python
import pandas as pd
````

````python
url = "https://en.wikipedia.org/wiki/List_of_Presidents_of_the_United_States"
````

````python
def getPresidents(url):
    urlreq = requests.get(url)
    urltext = urlreq.text
    expression = re.compile(r'title="Presidency of+.+">\d\d\d\d+.')
    matches = expression.findall(urltext)
    nameonly = re.compile(r'Presidency of (.+?)"')
    List_of_Presidents=list()
    for i in matches:
        name=nameonly.findall(i)
        List_of_Presidents.append(name[0])
    return List_of_Presidents
````

````python
def getBirthdays():
    Birthday_List=list()
    for i in getPresidents(url):
        x=i.replace(" ", "_")     ## to replace whitespaces in the name to 'underscore' as it is the url format
        urlreq = requests.get("https://en.wikipedia.org/wiki/"+x)
        urltext = urlreq.text
        pattern = re.compile(r'\d\d\d\d+-+\d\d+-+\d\d')
        matches = pattern.findall(urltext)
        Birthday_List.append(matches[0])
    return pd.Series(getPresidents(url),Birthday_List)
    
````

````python
B=[]
op_time_np=list()
````

````python
for i in range(100,250,10):
    B=np.random.rand(i)
    start=time.time()
    np.sort(B)
    end=time.time()
    elapsed=end-start
    op_time_np.append(elapsed)
````

##### Sorting using standard python statements

````python
def my_sort(B):
    for j in range(0,len(B)):
        for i in range(0,len(B)-1):
            temp=0
            if B[i]>B[i+1]:
                temp=B[i]
                B[i]=B[i+1]
                B[i+1]=temp
    return B



    
````

````python
op_time_std=list()

for i in range(100,250,10):
    B=np.random.rand(i)
    start=time.time()
    my_sort(B)
    end=time.time()
    elapsed=end-start
    op_time_std.append(elapsed)
    
````

````python
plt.plot(range(100,250,10),op_time_np)
plt.plot(range(100,250,10),op_time_std)
````

### Interpretation of the plot:

##### Numpy's sorting function uses quick sort for lower order arrays. Time complexity is O(n log(n)) in the best case and in the worst case it is O(n^2). It switches to heap sort for higher order to maintain the time complexity as O(n log(n).The sorting function using standard python statements has a time complexity of O(n^2).

## Part b) Determinant

##### Numpy determinant

````python
det_time_np=list()
````

````python
for i in range(2,20,1):
    A=np.random.randint(i*i,size=(i,i))
    start=time.time()
    np.linalg.det(A)
    end=time.time()
    elapsed=end-start
    det_time_np.append(elapsed)
    
    
    
    
````

##### Determinant using standard python statements

````python
def det(A,n):
    if len(A) == 1:
        determinant = ((n * A[0][0])*2)/10
        return determinant
    sign = 1
    sumCoeff = 0
    for i in range(len(A)):
        newMat = []
        for j in range(1, len(A)):
            coeff = []
            for k in range(len(A)):
                if k != i:
                    coeff.append(A[j][k])
            newMat.append(coeff)
        sign = -sign 
        sumCoeff = sumCoeff + n * det(newMat, sign * A[0][i])
    return sumCoeff
````

````python
det_time_std=list()
````

````python
for i in range(2,10,1):
    A=np.random.randint(i*i,size=(i,i))
    start=time.time()
    det(A,i)
    end=time.time()
    elapsed=end-start
    det_time_std.append(elapsed)
````

#### Blue : using numpy's determinant                           
#### Orange : using standard python statements

````python
plt.plot(range(2,20,1),det_time_np)
plt.plot(range(2,20,1),det_time_std)
````

### Interpretation of the plot:

##### Numpy's Determinant uses LU decomposition whose time complexity is O(n^3). The determinant function built using standard python functions is of time complexity O(n!).

### part c) Matrix Multiplication

````python
mul_time_np=list()
````

````python
for i in range(100,150):
    A=np.random.randint(i*i,size=(i,i))
    start=time.time()
    np.matmul(A,A)
    end=time.time()
    elapsed=end-start
    mul_time_np.append(elapsed)    
````

##### Matrix Multiplication using standard python statements

````python
mul_time_std=list()
````

````python
def matmult(a,b):
    zip_b = zip(*b)
    zip_b = list(zip_b)
    return [[sum(ele_a*ele_b for ele_a, ele_b in zip(row_a, col_b)) 
             for col_b in zip_b] for row_a in a]
````

````python
for i in range(100,150):
    A=np.random.randint(i*i,size=(i,i))
    start=time.time()
    matmult(A,A)
    end=time.time()
    elapsed=end-start
    mul_time_std.append(elapsed)
````

````python
plt.plot(range(100,150),mul_time_np)
plt.plot(range(100,150),mul_time_std)
````

### Interpretation of the plot:

##### Numpy uses highly optimized and carefully tuned BLAS ( basic Linear Algebra Subprograms) for matrix multiplication. The matrix multiplication function using standard python statements is of complexity O(n^3)
